```python
import re
#Define a function to standardize locations
def standardize_location(location):
    if pd.isna(location):
        return 'Unknown'
    location = str(location).strip()  # Remove extra spaces
    if re.match(r'http://|https://|www\.', location):
        return 'Unknown'
    location = location.replace('New York, NY', 'New York')
    location = location.replace('Seattle duh!', 'Seattle').replace('Grand Rapids,MI & Orlando, FL', 'Orlando')
    location = location.replace('South O-H-I-O Side', 'Ohio')
    location = location.replace('~', '').replace('(', '').replace(')', '')  # Remove special characters
    return location  
```

```python
# Apply the function to the DataFrame
df['standardized_location'] = df['tweet_location'].apply(standardize_location)
df.head()
```

```python
df.standardized_location.value_counts()
```

```python
df = df.drop('tweet_location', axis = 1)
df.head()
```

```python
#Define the mapping dictionary
negativereason_dict = { 
    'Customer Service Issue': '1',
    'Late Flight': '2',
    'Can\'t Tell': '3',
    'Cancelled Flight': '4',
    'Lost Luggage': '5',
    'Bad Flight': '6',
    'Flight Booking Problems': '7',
    'Flight Attendant Complaints': '8',
    'Long Lines': '9',  # Corrected key
    'Damaged Luggage': '10'
}

# Convert to integer type
df['negativereason'] = df['negativereason'].astype('str')

# Map the negativereason values
df['negativereason'] = df['negativereason'].map(negativereason_dict)
```

```python
# Replace NaN/empty values with '0'
df['negativereason'] = df['negativereason'].fillna('0')

# Convert to integer type if needed
df['negativereason'] = df['negativereason'].astype(int)

```

```python
df['negativereason_confidence'] = df['negativereason_confidence'].fillna('0')
df.head()
```

```python
df['negativereason_confidence'] = pd.to_numeric(df['negativereason_confidence'], errors='coerce')
```

```python
df['negativereason_confidence'] = df['negativereason_confidence'].round(2)
df.head()
```

```python
df['airline_sentiment_confidence'] = pd.to_numeric(df['airline_sentiment_confidence'], errors='coerce')
```

```python
df['airline_sentiment_confidence'] = df['airline_sentiment_confidence'].round(2)
df.head()
```

```python
df_col_reorder = [ 'name', 'tweet_date', 'standardized_location', 'airline', 'airline_sentiment', 'airline_sentiment_confidence', 'negativereason', 'negativereason_confidence']
df_col_reorder

```
